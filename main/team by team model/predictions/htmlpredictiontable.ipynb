```python
import pandas as pd
```

```python
# import our input datasets
df_predict = pd.read_csv('Prediction_Table.csv')
df_predict.head()
```

```python
predicted_df = df_predict.set_index("Team")
predicted_df.head()
```

```python
#rename the columns
predicted_df = predicted_df.rename(columns={"W/L Score":"W/L%"})
predicted_df = predicted_df.rename(columns={"R-squared training":"R^2 Train"})
predicted_df = predicted_df.rename(columns={"R-squared testing":"R^2 Test"})
predicted_df.head()
```

```python
#convert to html
tbtptable = predicted_df.to_html()
tbttable = tbtptable.replace('\n','')
tbttable
```

```python
#Get all the good scores
goodpredict_df = predicted_df.loc[predicted_df["R^2 Test"] > 0]
goodpredict_df
```

```python
#convert to html
tbtp2table = goodpredict_df.to_html()
tbt2table = tbtp2table.replace('\n','')
tbt2table
```
