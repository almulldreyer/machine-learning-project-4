# Data pull

```python
# Make a list of all team names and make a list for table names
path = 'C:\\Users\\kounn\\OneDrive\\Documents\\Bootcamp_Project4\\machine-learning-project-4\\data\\'
teams = ['ATL','BOS','CHA','CHI','CLE','DAL','DEN','DET','GSW','HOU','IND','LAC','LAL','MEM','MIA','MIL','MIN','NJN','NOH','NYK','OKC','ORL','PHI','PHO','POR','SAC','SAS','TOR','UTA','WAS']
winloss_teams = ['ATL','BOS','CHA','CHI','CLE','DAL','DEN','DET','GSW','HOU','IND','LAC','LAL','MEM','MIA','MIL','MIN','NJN','NOH','NYK','OKC','ORL','PHI','PHO','POR','SAC','SAS','TOR','UTA','WAS']
table_name = []
table_path = []
winloss_tables = []
team_tables = []

# Create CSV names
for team in teams:
    table_name.append(f"table_{team}.csv")

# Create Pathways
for table in table_name:
    table_path.append(path + str(table))
```

```python
# Put all stats tables into a list


# Create URL for all queries
for team in teams:
    query_url = f"https://www.basketball-reference.com/teams/{team}/stats_basic_yr_yr.html"
    df = pd.read_html(query_url)
    
    # Append each table into team_tables
    team_tables.append(df[0])    
        
```

```python
# Put all win/loss tables into a list
# Create URL for all queries
for team2 in winloss_teams:
    query_winloss = f"https://www.basketball-reference.com/teams/{team2}/"
    df_winloss = pd.read_html(query_winloss)
    
    # Append each table into winloss_tables
    winloss_tables.append(df_winloss[0])        
```

# Data cleaning

```python
# Drop NAN columns for all tables
x = 0

while x < len(teams):
    team_tables[x] = team_tables[x].drop(columns=['Unnamed: 3', 'Unnamed: 7', 'Season', 'Lg', 'Tm', 'Age', 'Ht.', 'Wt.'])
    x = x+1
```

```python
# Add "Win-Loss%" to "team_tables"
x = 0

while x < len(teams):
    team_tables[x]['W/L%'] = winloss_tables[x]['W/L%']
    x = x+1
```

```python
team_tables[1]
```

```python
# Drop all NAN rows for all tables
x = 0

while x < len(teams):
    team_tables[x] = team_tables[x].dropna()
    team_tables[x] = team_tables[x][team_tables[x].G != 'G']
    x = x+1
```

```python
team_tables[2]
```

```python
# Change all dtypes to fit ML Model
x = 0

while x < len(teams):
    team_tables[x]['G'] = team_tables[x]['G'].astype('int')
    team_tables[x]["MP"] = team_tables[x]["MP"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["FG"] = team_tables[x]["FG"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["FGA"] = team_tables[x]["FGA"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["3P"] = team_tables[x]["3P"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["3PA"] = team_tables[x]["3PA"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["2P"] = team_tables[x]["2P"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["FT"] = team_tables[x]["FT"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["FTA"] = team_tables[x]["FTA"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["ORB"] = team_tables[x]["ORB"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["DRB"] = team_tables[x]["DRB"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["TRB"] = team_tables[x]["TRB"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["AST"] = team_tables[x]["AST"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["STL"] = team_tables[x]["STL"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["BLK"] = team_tables[x]["BLK"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["TOV"] = team_tables[x]["TOV"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["PF"] = team_tables[x]["PF"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["PTS"] = team_tables[x]["PTS"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["2PA"] = team_tables[x]["2PA"].str.rstrip('%').astype('float') / 100.0
    team_tables[x]["FG%"] = team_tables[x]["FG%"].astype('float')
    team_tables[x]["3P%"] = team_tables[x]["3P%"].astype('float')
    team_tables[x]["2P%"] = team_tables[x]["2P%"].astype('float')
    team_tables[x]["FT%"] = team_tables[x]["FT%"].astype('float')
    x = x+1
```

# Data export

```python
# Export all DF to Data folder
x = 0

while x < len(teams):
    pd.DataFrame.to_csv (team_tables[x] ,table_path[x], index = False, header=True)
    x = x+1
```
