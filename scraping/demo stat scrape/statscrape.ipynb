```python
#Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import requests
```

```python
#Splinter 
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#Add in url and run browser
url = 'https://www.basketball-reference.com/teams/ATL/2022.html'
browser.visit(url)
html=browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#find first table
tables = pd.read_html(url)
atltables = tables[1]
atltables
```

```python
#just to check what was in the table that wasnt showing
atltables.columns
```

```python
#clean table
atl = atltables.set_index("Rk")
atl1 = atl.rename(columns={'Unnamed: 1':'Player'})
atl2 = atl1.dropna(subset=['Player'])
atl2
```

```python
atl2 = atl2.drop(columns={"GS", "FG", "FGA", "3P", "3PA", "2P", "2PA", "FT", "FTA"})
atl2
```

```python
#convert to html format
ATLtable = atl2.to_html()
ATLtable
```

```python
#clean html format
ATLt = ATLtable.replace('\n','')
ATLt
```
